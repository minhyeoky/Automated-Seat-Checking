# 1. ~87.xlsx

```python
data = pd.read_excel('../label/1-87_clean_2.xlsx', index_col=0)
data.head()
```

```python
data.drop(columns=['특이사항'], inplace=True)

data.drop(labels=1, inplace=True)
```

```python
t = data.values
```

```python
t[0]
```

```python
# \n 제거 및 ? => 0으로 통일
labels = ['[]', '[]']

for i in range(len(t)):
    labels.append(t[i][0])

for i in range(len(labels)):
    labels[i] = labels[i].replace('\n', '').replace('?', '0')
```

```python
labels[9]
```

```python

dummy = np.array([[None]]) # 인덱스 헷갈려서

real_labels = []
for i in range(len(labels)):
    print(labels[i])
    real_labels.append(np.array(ast.literal_eval(labels[i])))
```

# 2. 88~.txt


```python
data = []
with open('../label/88~_clean_2.txt', mode='r', encoding='utf8') as f:
    for line in f:
        data.append(line.strip('\n').split(' '))
```

```python
data = data[:-3]
```

```python
for each in data:
    temp = ast.literal_eval(each[1])
    real_labels.append(np.array(temp))
```

```python
len(real_labels)
```

```python
for i in range(len(real_labels)):
    if real_labels[i].shape[0] == 7:
        if real_labels[i].shape[1] == 3:
            real_labels[i] = np.concatenate((real_labels[i], [[0,0,0]]), axis=0)
```

```python
# 저장
with open('../label/clean_2_label.pickle', mode='wb') as f:
    pickle.dump(real_labels, f)
```

## reversed

```python
real_labels_r = []
real_labels = np.array(real_labels)
for i in range(len(real_labels)):
    r = real_labels[i]
    if len(real_labels[i].shape) is 2:
        r = np.flip(r, axis=1)
    real_labels_r.append(r)
```

```python

with open('../label/clean_2_label_r.pickle', mode='wb') as f:
    pickle.dump(real_labels_r, f)
```
